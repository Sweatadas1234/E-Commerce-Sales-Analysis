```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')

# Extract year, month, and day for analysis
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

# Example: Group by Year and Month to see monthly sales
monthly_sales = df.groupby(['Year', 'Month'])['Amount'].sum()
print(monthly_sales)
```

```python
# Example: Total sales by city
sales_by_city = df.groupby('ship-city')['Amount'].sum()
print(sales_by_city)

# Example: Total sales by state
sales_by_state = df.groupby('ship-state')['Amount'].sum()
print(sales_by_state)
```

```python
# Example: Total sales B2B vs B2C
b2b_sales = df[df['B2B'] == True]['Amount'].sum()
b2c_sales = df[df['B2B'] == False]['Amount'].sum()

print(f'B2B Sales: {b2b_sales}')
print(f'B2C Sales: {b2c_sales}')
```

```python
# Example: Total sales by city
sales_by_city = df.groupby('ship-city')['Amount'].sum()

# Get the top 10 cities by total sales
top_10_cities = sales_by_city.nlargest(10)
print("Top 10 Cities by Total Sales:\n", top_10_cities)
```

```python
# Example: Count of returns and cancellations
returns_cancellations = df[df['Status'].str.contains('Cancelled|Returned')].groupby('Status').size()
print(returns_cancellations)

# Example: Amount lost due to cancellations
amount_lost = df[df['Status'] == 'Cancelled']['Amount'].sum()
print(f'Amount lost due to cancellations: {amount_lost}')
```
